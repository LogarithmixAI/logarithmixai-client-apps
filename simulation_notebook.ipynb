# Traffic Simulation Notebook

## Purpose
This Jupyter Notebook is designed to simulate traffic patterns in a client-server architecture using the `TrafficSimulator` class from the `traffic_simulator` module. The simulation aims to model user interactions, request rates, and load profiles to analyze performance under various conditions such as normal, peak, and burst loads.

### Key Features:
- **Client-Server Setup**: Simulates interactions with a server running on `http://127.0.0.1:3000`.
- **Configurable Parameters**: Allows setting the number of users, simulation duration, maximum concurrency, debug mode, user request rate, and a custom load profile.
- **Load Profile**: Defines phases of the simulation, including normal operation, peak loads, bursts, and cooldown periods, to mimic real-world traffic scenarios.
- **Debugging**: Includes debug options for monitoring and troubleshooting the simulation process.

This notebook serves as a tool for developers and engineers to test and optimize server performance, identify bottlenecks, and ensure scalability in a controlled environment. The simulation can be run multiple times with different parameters to compare results and refine system configurations. -->

```python
from traffic_simulator import TrafficSimulator
sim = TrafficSimulator(
     base_url="http://127.0.0.1:3000",
    users=4,
    duration=30,
    max_concurrency= 2,
    # debug=False,
    debug=True,
    user_request_rate=2,
    load_profile= [
        (30, 2),   # normal
        (40, 11),  # peak
        (20, 5),  # peak
        (30, 3),  # peak
        (10, 1), # burst
        (40, 3)    # cooldown
    ],
)

sim.start()
```
